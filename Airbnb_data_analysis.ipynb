```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Check Missing Value

```python
df.isnull().sum()
```

```python
df.drop(columns=['house_rules', 'license'],inplace = True)
```

```python
df.head()
```

```python
# change the format of last review columns object to date format

df['last review'] = pd.to_datetime(df['last review'],errors='coerce')
```

```python
print(df['last review'])
```

# handling missing values


```python
df.fillna({'reviews per month':0 , 'last review':df['last review'].min()},inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# drop missing name and host name rows

df.dropna(subset=['NAME','host name'],inplace=True)
```

```python
# correct data types of price and service price
df['price'] = df['price'].replace('[\$,]','',regex = True).astype(float)
df['service fee'] = df['service fee'].replace('[\$]','',regex=True).astype(float)
```

```python
df.head()
```

```python
# check duplicates
df.drop_duplicates(inplace= True)
```

# data cleaning check

```python
df.info()
```

```python
# descriptive statistics
df.describe()
```

# 1. What is the distribution of listing prices?

```python
plt.figure(figsize=(10,6))
plt.title("dristibution of listing price")
plt.xlabel("price")
plt.ylabel('frequency')
sns.histplot(df['price'],bins=50,kde=True,color='red')
plt.show()
```

# 2. How are different room types distributed?


```python
plt.figure(figsize=(8,5))
plt.title("room type distribution")
plt.xlabel("room type")
plt.ylabel('count')
sns.countplot(data =df, x = 'room type',color= "hotpink")
plt.show()
```

# 3. How are listings distributed across neighborhoods?


```python
plt.figure(figsize=(12,8))
plt.title("listings distribution across neighborhoods")
plt.xlabel("neighborhood")
plt.ylabel('count')
sns.countplot(data=df, y = 'neighbourhood group',color='green', order = df['neighbourhood group'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

# 4. What is the relationship between price and room type?


```python
plt.figure(figsize = (10,6))
plt.title("price vs room type")
plt.xlabel("Room Type")
plt.ylabel("$Price")
sns.boxplot(data = df,x = 'room type',y = 'price', hue = 'room type', palette = 'Set1')
plt.show()
```

# 5. How has the number of reviews changed over time?

```python
review_over_time = df.groupby(df['last review'].dt.to_period('M')).size()
review_over_time.plot(kind='line',color ='red' ,figsize=(12,6))
plt.title("Number of Reviews Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Reviews")
plt.show()
```
